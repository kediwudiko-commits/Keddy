## Plotting the Keeling curve
Monthly mean carbon dioxide has been measured at Mauna Loa Observatory, Hawaii, since 1958.  The carbon dioxide data on Mauna Loa constitute the longest record of direct measurements of CO$_2$ in the atmosphere.

This record is referred to as the *Keeling curve* in honour of Charles Keeling who started the monitoring program and ran it until his death in 2005.

1. Search for the Mauna Loa monthly CO$_2$ data.
2. Download the data and put it into a Excel workbook (or similar)
3. Create a new Colab notebook (you don't need a GitHub template for this).
4. "Mount" the data in your Colab notebook
5. Use pandas to read in the CO$_2$ data.
6. Use Matplotlib to plot the CO$_2$ data against time, label your axes, etc.

```python
import pandas as pd
import matplotlib.pyplot as plt
# Load the dataset
# If it's an Excel file:
df = pd.read_excel('co2_mm_mlo.xlsx',skiprows=40)
# Check the first few rows to identify column names
print(df.head())
```

```python
# 1. Remove missing values (NOAA uses -99.99 for missing data)
df_clean = df[df['average'] > 0]

# 2. Create the plot
plt.figure(figsize=(12, 6))

# Use 'decimal date' for the x-axis and 'average' for the y-axis
plt.plot(df_clean['decimal date'], df_clean['average'], color='red', linewidth=1)

# Add title and labels as required by Step 6
plt.title('The Keeling Curve: Monthly Mean CO2 at Mauna Loa', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('CO2 Concentration (ppm)', fontsize=12)

# Add a grid for better readability
plt.grid(True, linestyle='--', alpha=0.6)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Ensure the data is sorted by date and filtered
df_monthly = df[df['average'] > 0].sort_values('decimal date')

# 2. Create the Monthly Plot
plt.figure(figsize=(14, 7))

# Plot with a thinner line and small markers to show monthly data points
plt.plot(df_monthly['decimal date'], df_monthly['average'],
         color='Red', linewidth=0.8, marker='o', markersize=2, label='Monthly Mean CO2')

# Step 6: Labels and Title
plt.title('Monthly Atmospheric CO2 Concentration at Mauna Loa', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('CO2 Concentration (ppm)', fontsize=12)

# --- OPTIONAL: ZOOM IN TO SEE MONTHLY VARIATION ---
# Uncomment the line below to see a specific 5-year window (e.g., 2015-2020)
# plt.xlim(2015, 2020)

# Add a legend and grid
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)

# Show the plot
plt.show()
```
